```python
def cosine_similarity_single(query, document):
    query_vec = Counter(query.lower().split())
    doc_vec = Counter(document.lower().split())
    
    # Common words
    common = set(query_vec.keys()) & set(doc_vec.keys())
    
    # Dot product
    dot_product = sum(query_vec[word] * doc_vec[word] for word in common)
    
    # Magnitudes
    mag_query = math.sqrt(sum(val**2 for val in query_vec.values()))
    mag_doc = math.sqrt(sum(val**2 for val in doc_vec.values()))
    
    if mag_query == 0 or mag_doc == 0:
        return 0
    
    return dot_product / (mag_query * mag_doc)
```

```python
document = "Rainfall helps crops grow and is important for agriculture"
query = "Why is rain important for farming?"
```

```python
cosine_similarity_single(query, document)
```
